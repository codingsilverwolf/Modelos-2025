```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

# Define the function for the right-hand side of the ODE
def f(x, y):
    return -2*x*y + 1

# Set the time span and initial condition
t_span = (0, 10)  # Integrate from t=0 to t=10
y0 = [0]         # Initial condition y(0) = 1

# Specify the time points where the solution is evaluated
t_eval = np.linspace(t_span[0], t_span[1], 100)

# Solve the initial value problem
solution = solve_ivp(f, t_span, y0, t_eval=t_eval, dense_output = True)

# Plot the results
plt.plot(solution.t, solution.y[0])
plt.xlabel('t')
plt.ylabel('y(t)')
plt.title('Solution to dy/dt = -0.5y')
plt.grid(True)
plt.show()
```
