```python
import numpy as np
import pandas as pd

# Número de muestras (horas)
n_samples = 60

# Creo un numpy array con las horas 1, 2, 3, ...
horas = np.arange(1, n_samples + 1)

base = 1.15  # Base de la función exponencial (crecimiento de E. coli)
scale = 20  # Escala del ruido
poblacion_inicial = 20  # Población inicial
poblacion_minima = 15  # Población mínima permitida

# Esta es la población sin ruido para las horas 1, 2, 3, ...
poblacion = poblacion_inicial * (base ** horas)
poblacion = np.round(poblacion).astype(int)  # Redondear la población a enteros

# Generar datos exponenciales con ruido
np.random.seed(0)  # Establezco una semilla fija para reproducibilidad

# Generar ruido normal con una desviación estándar que sigue a noise_scale
initial_noise_scale = 15  # Escala inicial del ruido
final_noise_scale = 9000  # Escala final del ruido
noise_scale = np.geomspace(initial_noise_scale, final_noise_scale, n_samples)  # Escala del ruido que aumenta gradualmente

# Generar ruido normal con una desviación estándar que varía con el tiempo
noise = np.random.normal(0, 1, size=poblacion.shape) * noise_scale
noise = np.round(noise).astype(int)  # Redondear el ruido a enteros
poblacion_noisy_1 = poblacion + noise

# Asegurar que la población con ruido no tenga valores negativos ni llegue a cero
poblacion_noisy_1 = np.maximum(poblacion_noisy_1, poblacion_minima)

# Seleccionar un subconjunto de la población para agregar más ruido
subset_indices = np.random.choice(np.arange(n_samples), size=int(n_samples * 0.5), replace=False)
additional_noise_scale = np.geomspace(initial_noise_scale, 0.25*(final_noise_scale), len(subset_indices))  # Escala del ruido adicional que aumenta gradualmente
additional_noise = np.random.normal(0, 1, size=subset_indices.shape) * additional_noise_scale
additional_noise = np.round(additional_noise).astype(int)  # Redondear el ruido adicional a enteros

# Crear una copia de poblacion_noisy_1 para agregar el ruido adicional
poblacion_noisy_2 = poblacion_noisy_1.copy()

# Agregar el ruido adicional al subconjunto seleccionado
poblacion_noisy_2[subset_indices] += additional_noise

# Asegurar que la población con ruido adicional no tenga valores negativos ni llegue a cero
poblacion_noisy_2 = np.maximum(poblacion_noisy_2, poblacion_minima)

# Crear el dataframe
df_E_Coli = pd.DataFrame({
    'Horas': horas,
    'Población': poblacion_noisy_2,
})

# Guardar el dataframe en un archivo CSV
df_E_Coli.to_csv('E_Coli.csv', index=False)

df_E_Coli
```
